```python
#!/big/venv/cuda/bin/python
from langchain_gigachat.chat_models import GigaChat
import speech_recognition as sr
import configparser


from gtts import gTTS
import os

from langchain_core.messages import HumanMessage, SystemMessage


config = configparser.ConfigParser()
config.read('config.ini')
API_KEY=config.get('gigachat', 'authkey')

giga = GigaChat(
    credentials=API_KEY,
    verify_ssl_certs=False,
)


def ask_model(msg, promt="Переведи следующее сообщение с русского на английский"):
    messages = [
        SystemMessage(content=promt),
        HumanMessage(content=msg),
    ]
    ans = giga.invoke(messages)
    return ans.content

def text_to_speech(text, out_filename='output.mp3', play = False):
    tts = gTTS(text=text, lang='ru')
    tts.save(out_filename)
    if play:
        os.system(f'mpg321 {out_filename}')  # Воспроизведение файла с использованием mpg



def file_to_text(filename='in.wav'):
    r = sr.Recognizer()
    audio = sr.AudioData.from_file(filename)
    said = ""
    try:
        said = r.recognize_google(audio, language = 'ru-RU')
    except Exception as e:
        print("Exception: + " + str(e))
    return said

def record_mic_file(filename='in.wav'):
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('Говорите')
        audio = r.listen(source)
        
    # write audio to a WAV file
    with open(filename, "wb") as f:
            f.write(audio.get_wav_data())        
```

```python
def main():
    filename='in2.wav'
    record_mic_file(filename)
    said = file_to_text(filename)
    print(f'Q:{said}')
    ans = ask_model(said,promt='Представь, что ты эксперт HR отдела. Задавай вопросы и отвечай без символов форматирования')
    ans = ans.replace('*','')
    print(f'A:{ans}')
    text_to_speech(ans,play=True)
    
    
```

```python
main()
```
